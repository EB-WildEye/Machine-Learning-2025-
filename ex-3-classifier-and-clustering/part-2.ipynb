# K-Means Clustering
calculation example of k-means clustering 

first lets define the points:


p_1 = (1, 2)

p_2 = (2, 1) 

p_3 = (2.5, 2.5) 

p_4 = (4, 2)

p_5 = (5, 1) 

p_6 = (5, 3)

p_7 = (6.5, 2)

### Assign initial centroids

centroids:

c_1 = (2, 3)

c_2 = (4, 3)

### Assigment step - step 1 ~ 
points assignment to clusters with auclidean distance:

argmin{ || p_i - c_j || } for each point p_i, c_j, when i = 1, ..., 7 and j = 1, 2

.

c_1: ((1-2)^2 + (3-4)^2)^(1/2) = 1.414;   

c_2: ((1-4)^2 + (3-4)^2)^(1/2) = 3.162;

p_1  -->  assigned to c_1

.

c_1: ((2-2)^2 + (1-3)^2)^(1/2) = 2;

c_2: ((2-4)^2 + (1-4)^2)^(1/2) = 3.606;

p_2  -->  assigned to c_1

.

c_1: ((3-2)^2 + (5-3)^2)^(1/2) = 2.236;

c_2: ((3-4)^2 + (5-4)^2)^(1/2) = 1.414;

p_3  -->  assigned to c_1

.

c_1: ((4-2)^2 + (2-3)^2)^(1/2) = 2.236;

c_2: ((4-4)^2 + (2-4)^2)^(1/2) = 2;

p_4  -->  assigned to c_2

.

c_1: ((5-2)^2 + (1-3)^2)^(1/2) = 3.606;

c_2: ((5-4)^2 + (1-4)^2)^(1/2) = 3.162;

p_5  -->  assigned to c_2

.

c_1: ((6-2)^2 + (3-3)^2)^(1/2) = 4;

c_2: ((6-4)^2 + (3-4)^2)^(1/2) = 2.236;

p_6  -->  assigned to c_2

.

c_1: ((7-2)^2 + (2-3)^2)^(1/2) = 5.099;

c_2: ((7-4)^2 + (2-4)^2)^(1/2) = 3.606;

p_7  -->  assigned to c_2

### Centroid update - step 2 ~
c_1 = 1/2((2, 3) + (4, 3)) = (3, 3)

c_2 = 1/4((1, 2) + (2, 1) + (2.5, 2.5) + (4, 2) + (5, 1) + (5, 3) + (6.5, 2)) = (6.5, 3.375)


### Assigment step - step 2 ~
c_1: ((1-3)^2 + (2-3)^2)^(1/2) = 2.236;

c_2: ((1-6.5)^2 + (2-3.375)^2)^(1/2) = 5.669;

p_1  -->  assigned to c_2

.

c_1: ((2-3)^2 + (1-3)^2)^(1/2) = 2.236;

c_2: ((2-6.5)^2 + (1-3.375)^2)^(1/2) = 5.088;

p_2  -->  assigned to c_2

.

c_1: ((2.5-3)^2 + (2.5-3)^2)^(1/2) = 0.707;

c_2: ((2.5-6.5)^2 + (2.5-3.375)^2)^(1/2) = 4.094;

p_3  -->  assigned to c_1

.

c_1: ((4-3)^2 + (2-3)^2)^(1/2) = 1.414;

c_2: ((4-6.5)^2 + (2-3.375)^2)^(1/2) = 2.853;

p_4  -->  assigned to c_1

.

c_1: ((5-3)^2 + (1-3)^2)^(1/2) = 2.236;

c_2: ((5-6.5)^2 + (1-3.375)^2)^(1/2) = 2.809;

p_5  -->  assigned to c_1

.

c_1: ((5-3)^2 + (3-3)^2)^(1/2) = 2;

c_2: ((5-3.5)^2 + (3-2.5)^2)^(1/2) = 1.581;

p_6  -->  assigned to c_1

.

c_1: ((6.5-3)^2 + (2-3)^2)^(1/2) = 3.640;

c_2: ((6.5-3.5)^2 + (2-2.5)^2)^(1/2) = 3.041;

p_7  -->  assigned to c_1


### Step 2 ~ Error calculation:

first we need to calculate the centroid after step 2:

c_1: 1/5((1, 2) + (2, 1) + (2.5, 2.5) + (4, 2) + (5, 1)) = (2.9, 1.7)

c_2: 1/2((5, 3) + (6.5, 2)) = (5.75, 2.5)

.


now we can calculate the error:

E =  1/7 * (‖p₁–c₁‖² + ‖p₂–c₁‖² + ‖p₃–c₁‖² + ‖p₄–c₂‖² + ‖p₅–c₁‖² + ‖p₆–c₂‖² + ‖p₇–c₂‖²)

--> 1/7 * (3.70 + 1.30 + 0.80 + 1.30 + 4.90 + 0.8125 + 0.8125) = 13.625 / 7 ≈ 1.9464
