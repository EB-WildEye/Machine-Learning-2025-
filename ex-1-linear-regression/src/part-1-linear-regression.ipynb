```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

# Part 1 ~ Linear Regression

```python
# preparing the data
a1 = 1.8
a0 = -2
x = 10*np.random.rand(100)
y = a0 + a1*x +np.random.randn(100)
plt.scatter(x,y)

# using sklearn to fit the model
model = LinearRegression(fit_intercept=True)
model.fit(x[:,np.newaxis],y)
xfit = np.linspace(0,10,100) 
yfit = model.predict(xfit[:,np.newaxis])

plt.scatter(x,y)
plt.plot(xfit,yfit)

print("Model slope a1: ", model.coef_[0])
print("Model intercept a0: ", model.intercept_)
```

```python
x = np.random.rand(500) * 35
a = 2.7
b = 5
noise = np.random.normal(0, 25, 500)
y = a*x + b + noise

plt.scatter(x, y, color='purple')

model = LinearRegression()
model.fit(x.reshape(-1, 1), y)

x_fit = np.linspace(0, 35, 100)
y_fit = model.predict(x_fit.reshape(-1, 1))

plt.plot(x_fit, y_fit)
plt.show()

print("Model slope a1 = ", model.coef_[0])
print("Model intercept a0 = ", model.intercept_)
```
